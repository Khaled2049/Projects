```python
"""
Program: hw3.ipynb
Programmed by: Khaled Hossain
Description: An implementation of KNN algorithm 
"""
#---------------------------------Imports--------------------------------------
import numpy as np
import pandas as pd
from sklearn import preprocessing
from operator import itemgetter
#------------------------------------------------------------------------------

# Get the data
train_data = pd.read_csv('MNIST_train.csv')
test_data = pd.read_csv('MNIST_test.csv')
X_train = np.array(train_data)
X_test = np.array(test_data)



# Scale & normalize (gave worse accuracy)
def normalize(data):
    return preprocessing.normalize(data)
def scale(data):
    return preprocessing.scale(data)

#Calculates the distance between samples
def euclidean_distance(testData, trainData):
    distance = 0
    
    for i in range(len(testData)):
        distance += (testData[i] - trainData[i])**2
    
    return np.sqrt(distance)

# cosine_similarity (gave worse accuracy)
def cos_sim(a, b):

    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    return dot_product / (norm_a * norm_b)

#---------------------------------Classes/Functions----------------------------

class KNN:
    def __init__(self, k=4):
        self.k = k
    
    # Calculate Accuracy 
    def accuracy(self, testData, outputData):
        match = 0
        mistake = 0
        for i in range(len(testData)):
            if testData[i][0] == outputData[i]:
                match +=1
            else:
                mistake +=1
                
        accuracy = (match/len(outputData))*100
        return [accuracy, mistake]

    # get the weights for closest neighbors
    # compute the inverse of each distance, find the sum of the inverses, then divide each inverse by the sum
    def get_weights(self, data):

        for i in data:
            weight = 1/i[1]
            i.append(weight)
        
        #print(data)
        data.sort(key=itemgetter(2))
        #print(data)
        # now we have [Class, distance, weight]
        return data[0][0]
    
    
    
    # weighted K-nearest neighbor algorthim
    def weighted_knn(self, testData, trainData):
        
        outputData = []
        
        print('K = ', self.k)
        
        for data in testData:
            distance = []
            
            # using euclidean distance
            for t_data in trainData:
                distance.append([t_data[0], euclidean_distance(data[1:], t_data[1:])])
            distance.sort(key=itemgetter(1))
            
            # using cosine similarity
            #for train in trainData:
            #    distance.append([train[0], cos_sim(data[1:], train[1:])])
            #distance.sort(key=itemgetter(1))
            
            neighbors = []
            
            for i in range(self.k):
                neighbors.append(distance[i])
            
            result = self.get_weights(neighbors)
            outputData.append(result)
            #print(outputData)

            print('Desired Class = ', data[0], "Computed Class = ", result)

        # Display accuracy, misclassifications and total number of test samples
        accuracy_result = self.accuracy(testData, outputData)
        print ('Accuracy Rate = ', accuracy_result[0], "%")
        print('Number of misclassified test samples = ', accuracy_result[1])
        print('Total number of test samples = ', len(testData))





#noramlized data
# a = normalize(X_train)
# b = normalize(X_test)
# clf = KNN(k = 4)
# clf.weighted_knn(b, a)

# calling the classifier and setting the value of k to 4

clf = KNN(k = 4)

clf.weighted_knn(X_test, X_train)


```
